```python
from scipy.io import wavfile
from winsound import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read the wav file (mono)
samplingFrequency1, signalData1 = wavfile.read('misc/mix_female.wav')

plt.figure(1, figsize=(15,7))

# Plot the signal read from wav file
plt.figure(1)
plt.subplot(211)
plt.title('Input mix_female.wav file in time-domain, followed by Spectrograms')

plt.plot(signalData1)
plt.xlabel('Sample')
plt.ylabel('Amplitude')

plt.subplot(212)
plt.specgram(signalData1,Fs=samplingFrequency1)
plt.xlabel('Time')
plt.ylabel('Frequency')

plt.show()
```
